## Phase 2 - Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('Phase1_HealthCare.csv')
df.head()
```

## Count Plot

```python
sns.set_style('darkgrid')
sns.countplot(df['Outcome'])
plt.title("Countplot of Outcome")
plt.xlabel('Outcome')
plt.ylabel("Count")
print("Count of class is:\n",df['Outcome'].value_counts())
```

#### It is noticed that both class is balanced and beacause of this it is not required to perform any sampling method to maintain the balance between both classes. Therefore data will be directly used in in training and testing purpose without performing any sampling method. Also, for Model Validation ,  ROC Curve will not play major role because data is not imbalanced. However, data is related to medical segment,  ROC curve will be used to make sure no presence of TYPE 2 ERROR.

## Scatter Plot

```python
sns.pairplot(df)
plt.title('Scatter plot between variables')
```

#### With reference to scatter plot, below are observations

    2. However, positive correlation is possible between skin thickness and BMI, Pregnancies and age
